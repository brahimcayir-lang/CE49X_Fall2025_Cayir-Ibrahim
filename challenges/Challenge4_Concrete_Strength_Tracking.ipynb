# Challenge 4: Construction Quality Control

## Task: Track concrete strength development over time

### Requirements:
1. Create time data from day 1 to 28
2. Generate 3 concrete batches with different strength curves
3. Batch A: 25 + 0.5*day + noise
4. Batch B: 27 + 0.4*day + noise
5. Batch C: 23 + 0.6*day + noise
6. Use seaborn lineplot with confidence intervals
7. Add proper labels and legend

**First team to show the time series plot wins Challenge 4!** üèÜ


```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

print("üìö Libraries imported successfully!")
print("üîß Setting up concrete strength tracking system...")
```

```python
# 1. Create time data from day 1 to 28
days = np.arange(1, 29)
print(f"üìÖ Time data created: Days {days[0]} to {days[-1]}")

# 2-5. Generate 3 concrete batches with different strength curves
print("üèóÔ∏è Generating concrete batch data...")

# Batch A: 25 + 0.5*day + noise
batch_a_strength = 25 + 0.5 * days + np.random.normal(0, 0.8, len(days))

# Batch B: 27 + 0.4*day + noise  
batch_b_strength = 27 + 0.4 * days + np.random.normal(0, 0.6, len(days))

# Batch C: 23 + 0.6*day + noise
batch_c_strength = 23 + 0.6 * days + np.random.normal(0, 0.7, len(days))

print("‚úÖ All three batches generated successfully!")
print(f"   Batch A: Starting strength = {batch_a_strength[0]:.2f} MPa")
print(f"   Batch B: Starting strength = {batch_b_strength[0]:.2f} MPa")
print(f"   Batch C: Starting strength = {batch_c_strength[0]:.2f} MPa")
```

```python
# Create DataFrame for seaborn plotting
data = []
for i, day in enumerate(days):
    data.extend([
        {'Day': day, 'Batch': 'Batch A', 'Strength (MPa)': batch_a_strength[i]},
        {'Day': day, 'Batch': 'Batch B', 'Strength (MPa)': batch_b_strength[i]},
        {'Day': day, 'Batch': 'Batch C', 'Strength (MPa)': batch_c_strength[i]}
    ])

df = pd.DataFrame(data)
print("üìä DataFrame created for plotting")
print(f"   Shape: {df.shape}")
print(f"   Columns: {list(df.columns)}")
print("\nFirst few rows:")
print(df.head())
```

```python
# 6. Use seaborn lineplot with confidence intervals
# 7. Add proper labels and legend

plt.figure(figsize=(12, 8))

# Create the seaborn lineplot with confidence intervals
sns.lineplot(data=df, x='Day', y='Strength (MPa)', hue='Batch', 
             ci=95,  # 95% confidence interval
             marker='o', 
             markersize=6,
             linewidth=2.5)

# Add proper labels and legend
plt.title('Concrete Strength Development Over Time\nChallenge 4: Construction Quality Control', 
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Days', fontsize=14, fontweight='bold')
plt.ylabel('Compressive Strength (MPa)', fontsize=14, fontweight='bold')

# Customize legend
plt.legend(title='Concrete Batches', 
           title_fontsize=12, 
           fontsize=11,
           loc='lower right',
           frameon=True,
           fancybox=True,
           shadow=True)

# Add grid for better readability
plt.grid(True, alpha=0.3, linestyle='--')

# Set axis limits and ticks
plt.xlim(0, 29)
plt.xticks(range(0, 29, 3))  # Show every 3rd day
plt.ylim(20, 45)

# Add horizontal line for typical concrete strength threshold
plt.axhline(y=30, color='red', linestyle=':', alpha=0.7, 
            label='Typical Strength Threshold (30 MPa)')

# Adjust layout to prevent label cutoff
plt.tight_layout()

print("üìà Plot created successfully!")
print("üé® All requirements met:")
print("   ‚úÖ Time data from day 1 to 28")
print("   ‚úÖ 3 concrete batches with different strength curves")
print("   ‚úÖ Batch A: 25 + 0.5*day + noise")
print("   ‚úÖ Batch B: 27 + 0.4*day + noise")
print("   ‚úÖ Batch C: 23 + 0.6*day + noise")
print("   ‚úÖ Seaborn lineplot with confidence intervals")
print("   ‚úÖ Proper labels and legend")
```

```python
# Display the plot
plt.show()

# Print final summary
print("\n" + "="*60)
print("üèÜ CHALLENGE 4 COMPLETE! üèÜ")
print("="*60)
print("Final strengths at Day 28:")
print(f"   Batch A: {batch_a_strength[-1]:.2f} MPa")
print(f"   Batch B: {batch_b_strength[-1]:.2f} MPa") 
print(f"   Batch C: {batch_c_strength[-1]:.2f} MPa")

print(f"\nStrength growth rates:")
print(f"   Batch A: 0.5 MPa/day")
print(f"   Batch B: 0.4 MPa/day")
print(f"   Batch C: 0.6 MPa/day")

print("\nüéâ Time series plot generated successfully!")
print("üèÜ First team to show the time series plot wins Challenge 4!")
print("="*60)
```
